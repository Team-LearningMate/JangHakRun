```python
import pandas as pd

GRADE_TOKENS = [
    "대학2학기",
    "대학3학기",
    "대학4학기",
    "대학5학기",
    "대학6학기",
    "대학7학기",
    "대학8학기이상",
    "대학신입생",
    "석사신입생(1학기)",
    "석사2학기이상",
    "박사과정",
    "연령제한",
    "제한없음",
    "해당없음",
]

GRADE_MAPPING = {
    "대학2학기": "대학2학기",
    "대학3학기": "대학3학기",
    "대학4학기": "대학4학기",
    "대학5학기": "대학5학기",
    "대학6학기": "대학6학기",
    "대학7학기": "대학7학기",
    "대학8학기이상": "대학8학기이상",
    "대학신입생": "대학신입생",
    "석사신입생(1학기)": "석사신입생(1학기)",
    "석사2학기이상": "석사2학기이상",
    "박사과정": "박사과정",
    "연령제한": "연령제한",
    "제한없음": "제한없음",
    "해당없음": "해당없음",
}


MAJOR_TOKENS = [
    "공학계열",
    "교육계열",
    "사회계열",
    "예체능계열",
    "의약계열",
    "인문계열",
    "자연계열",
    "특정학과",
    "제한없음",
    "해당없음",
]

# 출력에서는 '계열' 떼고 저장
MAJOR_MAPPING = {
    "공학계열": "공학",
    "교육계열": "교육",
    "사회계열": "사회",
    "예체능계열": "예체능",
    "의약계열": "의약",
    "인문계열": "인문",
    "자연계열": "자연",
    "특정학과": "특정학과",
    "제한없음": "제한없음",
    "해당없음": "해당없음",
}


UNIV_TOKENS = [
    "4년제(5~6년제포함)",
    "기술대학",
    "원격대학",
    "전문대(2~3년제)",
    "일반대학원",
    "전문대학원",
    "학점은행제 대학",
    "해외대학",
    "특정대학",
    "제한없음",
    "해당없음",
]

UNIV_MAPPING = {
    "4년제(5~6년제포함)": "4년제(5~6년제포함)",
    "기술대학": "기술대학",
    "원격대학": "원격대학",
    "전문대(2~3년제)": "전문대(2~3년제)",
    "일반대학원": "일반대학원",
    "전문대학원": "전문대학원",
    "학점은행제 대학": "학점은행제 대학",
    "해외대학": "해외대학",
    "특정대학": "특정대학",
    "제한없음": "제한없음",
    "해당없음": "해당없음",
}

```

```python
def normalize_grade(text: str) -> str:
    """
    예)
    '대학2학기대학3학기대학4학기대학신입생'
      -> '대학2학기,대학3학기,대학4학기,대학신입생'
    """
    if pd.isna(text):
        return None

    found = []

    # 토큰 목록 순서대로 포함 여부 체크
    for raw in GRADE_TOKENS:
        if raw in text:
            found.append(GRADE_MAPPING[raw])

    # 중복 제거 + 원래 정의한 GRADE_TOKENS 순서 유지
    seen = set()
    ordered = []
    for tok in GRADE_TOKENS:
        if tok in found and tok not in seen:
            seen.add(tok)
            ordered.append(tok)

    return ",".join(ordered) if ordered else None

def normalize_major(text: str) -> str:
    """
    예)
    '공학계열교육계열자연계열' 
      -> '공학,교육,자연'
    """
    if pd.isna(text):
        return None

    found = []

    for raw in MAJOR_TOKENS:
        if raw in text:
            found.append(MAJOR_MAPPING[raw])

    # 토큰 순서 유지 + 중복 제거
    seen = set()
    ordered = []
    for raw in MAJOR_TOKENS:
        mapped = MAJOR_MAPPING[raw]
        if mapped in found and mapped not in seen:
            seen.add(mapped)
            ordered.append(mapped)

    return ",".join(ordered) if ordered else None

def normalize_univ_type(text: str) -> str:
    """
    예)
    '4년제(5~6년제포함)기술대학전문대(2~3년제)학점은행제 대학해외대학'
      -> '4년제(5~6년제포함),기술대학,전문대(2~3년제),학점은행제 대학,해외대학'
    """
    if pd.isna(text):
        return None

    found = []

    # 문자열에 포함된 토큰 찾아서 추가
    for raw in UNIV_TOKENS:
        if raw in text:
            found.append(UNIV_MAPPING[raw])

    # 정의된 토큰 순서를 기준으로 중복 제거
    seen = set()
    ordered = []
    for raw in UNIV_TOKENS:
        mapped = UNIV_MAPPING[raw]
        if mapped in found and mapped not in seen:
            seen.add(mapped)
            ordered.append(mapped)

    return ",".join(ordered) if ordered else None

```

```python
# 원본 '학년구분' 컬럼을 바로 정규화해서 덮어쓰기
df["학년구분"] = df["학년구분"].astype(str).apply(normalize_grade)
df["학과구분"] = df["학과구분"].astype(str).apply(normalize_major)
df["대학구분"] = df["대학구분"].astype(str).apply(normalize_univ_type)

df.head()
```

```python
df.to_csv("transformed_data.csv", index=False, encoding="utf-8-sig")
```
